```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
F: np.ndarray = np.random.normal(size=(100, 10))
F = (F - np.mean(F, axis=0)) / np.std(F, ddof=1)
pd.DataFrame(F)
```

```python
C: np.ndarray = np.dot(np.transpose(F), F)
pd.DataFrame(C)
```

```python
stock_cov: np.ndarray = np.dot(np.dot(F, C), np.transpose(F))
pd.DataFrame(stock_cov)
```
