### Arbitrage Detection and Risk-Neutral Bet Sizing

#### Arbitrage Detection

To detect arbitrage, the fundamental property we exploit is:

The implied probabilities \( P_i \) derived from the odds \( O_i \) for all \( i \) in a set of \( N \) bets should sum to 1 if the odds are fairly priced. That is:

$$
\sum_{i=1}^{N} P_i = 1 \quad \text{(fairly priced market)}.
$$

If the sum of the implied probabilities is less than 1, there is an arbitrage opportunity because the odds misprice the event, creating a potential payout greater than the total bet.

**Proof of Arbitrage Detection:**

Implied Probabilities: For odds \( O_i \), the implied probability is:

$$
P_i = \frac{1}{O_i + 1}.
$$

Arbitrage Condition: If:

$$
\sum_{i=1}^{N} P_i < 1,
$$

then the market is offering a "collective payout" for a single unit bet greater than the cost of placing proportional bets on all outcomes. This creates a guaranteed profit.

Mathematically:

$$
\text{Profit} = \text{Total Payout} - \text{Total Bet Amount} > 0.
$$

#### Risk-Neutral Bet Sizing

Let \( O_i \) be the odds for bet \(i\). The implied probability of each bet is:

$$
P_i = \frac{1}{O_i + 1}.
$$

Risk-neutral bet sizes ensure equal payouts across all outcomes. The bet size \(x_i\) for each bet is calculated to ensure:

$$
O_i \cdot x_i - \sum_{j \neq i} x_j = \text{constant across all bets}.
$$

Solving for bet sizes:

$$
x_i = \frac{K + C}{O_i + 1},
$$

where \(K\) is the total bankroll and \(C\) is a constant ensuring that the total allocation sums to \(K\). This leads to the final formula for each bet size:

$$
x_i = \frac{K}{(O_i + 1) \cdot \sum_{j=1}^{N} \frac{1}{O_j + 1}}.
$$

The bet size as a percentage of the bankroll is then:

$$
\text{weight}_i = \frac{x_i}{K} = \frac{1}{(O_i + 1) \cdot \sum_{j=1}^{N} \frac{1}{O_j + 1}}.
$$

#### Risk-Neutral Bet Sizing

To solve for **risk-neutral allocations** for an arbitrary number of mispriced bets \( N \), where the probabilities \( P_i \) derived from the odds don't sum to 1, we aim to ensure that the payouts are equal across all outcomes while respecting the capital allocation constraint. Here's how this can be approached mathematically:

---

### **1. Define Variables**
- Let \( N \) be the total number of bets.
- Let \( x_i \) be the amount bet on outcome \( i \) (\( i = 1, 2, \ldots, N \)).
- Odds for bet \( i \): \( O_i \), meaning the payout for \( i \) winning is \( O_i \cdot x_i \).
- Total capital available: \( K \).
- Implied probabilities from the odds: \( P_i = \frac{1}{O_i + 1} \).

---

### **2. Objectives**
1. **Payout equality**: The net payout should be the same regardless of which bet wins:
   $$
   O_1 \cdot x_1 - \sum_{j \neq 1} x_j = O_2 \cdot x_2 - \sum_{j \neq 2} x_j = \cdots = O_N \cdot x_N - \sum_{j \neq N} x_j.
   $$

2. **Capital constraint**: All bet allocations must sum to \( K \):
   $$
   \sum_{i=1}^N x_i = K.
   $$

---

### **3. Reformulate the Payout Equality**
From the payout equality condition:
$$
O_i \cdot x_i - \sum_{j \neq i} x_j = C, \quad \forall i,
$$
where \( C \) is the constant payout.

Rearrange:
$$
O_i \cdot x_i + x_i = \sum_{j=1}^N x_j + C \quad \Rightarrow \quad x_i(O_i + 1) = K + C.
$$

Solve for \( x_i \):
$$
x_i = \frac{K + C}{O_i + 1}. \tag{1}
$$

---

### **4. Use the Capital Constraint**
Substitute \( x_i = \frac{K + C}{O_i + 1} \) into the capital constraint:
$$
\sum_{i=1}^N x_i = K.
$$

$$
\sum_{i=1}^N \frac{K + C}{O_i + 1} = K.
$$

Factor \( K + C \) out of the summation:
$$
(K + C) \cdot \sum_{i=1}^N \frac{1}{O_i + 1} = K.
$$

Solve for \( K + C \):
$$
K + C = \frac{K}{\sum_{i=1}^N \frac{1}{O_i + 1}}. \tag{2}
$$

---

### **5. Solve for Each Bet Allocation**
Substitute \( K + C \) from (2) into (1):
$$
x_i = \frac{\frac{K}{\sum_{i=1}^N \frac{1}{O_i + 1}}}{O_i + 1}.
$$

Simplify:
$$
x_i = \frac{K}{(O_i + 1) \cdot \sum_{j=1}^N \frac{1}{O_j + 1}}. \tag{3}
$$

---

### **6. Interpretation**
1. \( x_i \) is proportional to the reciprocal of \( O_i + 1 \), normalized by the sum of all reciprocals of \( O_i + 1 \).
2. This ensures payouts are risk-neutral, as the proportional allocation guarantees identical payouts for all outcomes.

---

### **7. Systematic Algorithm**
1. Compute \( \frac{1}{O_i + 1} \) for each bet \( i \) (\( i = 1, \ldots, N \)).
2. Compute the normalization factor \( \sum_{i=1}^N \frac{1}{O_i + 1} \).
3. Allocate capital \( x_i \) using:
   $$
   x_i = \frac{K}{(O_i + 1) \cdot \sum_{j=1}^N \frac{1}{O_j + 1}}.
   $$

---

### **Example**
Suppose you have \( N = 3 \) bets:
- Odds \( O_1 = 2:1 \), \( O_2 = 3:1 \), \( O_3 = 4:1 \).
- Total capital \( K = 100 \).

1. Compute reciprocals of \( O_i + 1 \):
   $$
   \frac{1}{O_1 + 1} = \frac{1}{3}, \quad \frac{1}{O_2 + 1} = \frac{1}{4}, \quad \frac{1}{O_3 + 1} = \frac{1}{5}.
   $$

2. Sum these values:
   $$
   \sum_{i=1}^3 \frac{1}{O_i + 1} = \frac{1}{3} + \frac{1}{4} + \frac{1}{5} = \frac{20}{60} + \frac{15}{60} + \frac{12}{60} = \frac{47}{60}.
   $$

3. Compute each \( x_i \):
   $$
   x_1 = \frac{100}{(2 + 1) \cdot \frac{47}{60}} = \frac{100}{3 \cdot \frac{47}{60}} = \frac{100 \cdot 60}{141} \approx 42.55,
   $$
   $$
   x_2 = \frac{100}{(3 + 1) \cdot \frac{47}{60}} = \frac{100}{4 \cdot \frac{47}{60}} = \frac{100 \cdot 60}{188} \approx 31.91,
   $$
   $$
   x_3 = \frac{100}{(4 + 1) \cdot \frac{47}{60}} = \frac{100}{5 \cdot \frac{47}{60}} = \frac{100 \cdot 60}{235} \approx 25.53.
   $$

---

#### Risk-Neutral Bet Sizing Generalization
This method scales for any $ N $. The key is to:

1. Use $ \frac{1}{O_i + 1} $ for each bet.
2. Normalize by their sum.
3. Allocate capital proportionally.

This ensures **payout neutrality** across all outcomes.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def detect_arbitrage(odds: np.ndarray) -> bool:
    """
    Detect whether an arbitrage opportunity exists based on a set of mispriced odds.

    Parameters:
    odds (np.ndarray): Array of odds for each bet (e.g., [2, 3, 4] for 2:1, 3:1, 4:1 odds).

    Returns:
    bool: True if arbitrage opportunity exists, False otherwise.

    To detect arbitrage, we check if the sum of the implied probabilities is less than 1.

    The implied probability for each bet is calculated as:
    P_i = 1 / (O_i + 1)
    
    If the sum of the implied probabilities for all bets is less than 1, an arbitrage opportunity exists.

    Explanation of Arbitrage Detection:
    If the sum of the implied probabilities (P_i) is less than 1, there is an arbitrage opportunity because 
    the market is offering a "collective payout" that is greater than the cost of placing proportional bets 
    on all outcomes, which results in a guaranteed profit.
    """

    # Compute implied probabilities P_i = 1 / (O_i + 1)
    implied_probabilities = 1 / (odds + 1)

    # Check if the sum of probabilities is less than 1
    total_implied_probability = np.sum(implied_probabilities)

    return total_implied_probability < 1


def get_risk_neutral_bet_weights(odds: np.ndarray) -> np.ndarray:
    """
    Calculate risk-neutral bet sizes based on mispriced odds for an arbitrary number of bets.

    Parameters:
    odds (np.ndarray): Array of odds for each bet (e.g., [2, 3, 4] for 2:1, 3:1, 4:1 odds).

    Returns:
    np.ndarray: Bet sizes as a fraction of the bankroll for each bet.

    To calculate risk-neutral bet sizes, we use the following approach:

    1. Compute the implied probabilities for each bet as P_i = 1 / (O_i + 1).
    2. Normalize the probabilities by calculating their sum.
    3. The bet sizes are proportional to these normalized probabilities, ensuring that the total allocation sums to the bankroll.
    4. Each bet size is given by:
       x_i = (K / (O_i + 1)) / (sum(1 / (O_j + 1) for all j)),
       where K is the bankroll and O_i are the odds for each bet.

    This ensures the same payout across all bets, regardless of the odds.
    """

    # Compute implied probabilities P_i = 1 / (O_i + 1)
    implied_probabilities = 1 / (odds + 1)

    # Normalize implied probabilities
    normalization_factor = np.sum(implied_probabilities)

    # Calculate risk-neutral bet sizes as a percentage of bankroll
    weights = implied_probabilities / normalization_factor

    return weights
```

```python
def bet(odds: np.ndarray, bankroll: float) -> float:
    """
    Simulate a betting round with repeated bets.
    The outcome is randomly chosen based on the odds provided.
    """
    
    # Detect arbitrage opportunity
    arbitrage = detect_arbitrage(odds)
    
    if arbitrage:

        # Calculate risk-neutral bet sizes
        bet_weights = get_risk_neutral_bet_weights(odds) * bankroll
        
        # Log the detected arbitrage and bet sizes
        print(f"Arbitrage detected! Bet sizes: {bet_weights}")
        
        # Simulate the outcome (random choice based on odds)
        bet_outcome = np.random.choice(len(odds), p=(1 / (odds + 1)) / np.sum(1 / (odds + 1)))
        
        # Calculate the payout based on the winning bet
        payout = bet_weights[bet_outcome] * (odds[bet_outcome] + 1)
        bankroll = bankroll - np.sum(bet_weights) + payout
        
        # Log the outcome
        print(f"Outcome: Bet on market {bet_outcome + 1} wins (Odds: {odds[bet_outcome]})")
        print(f"Payout: ${payout:.2f}")
    
    else:
        
        print("No arbitrage opportunity detected.")
    
    return bankroll


def run_simulation(odds: np.ndarray, initial_bankroll: float, n_rounds: int):
    """
    Run a Monte Carlo simulation of repeated bets.
    """

    bankroll = initial_bankroll
    bankroll_history = [bankroll]

    for i in range(n_rounds):

        print(f"\nRound {i + 1} - Current Bankroll: ${bankroll:.2f}")
        
        # Perform the bet and update the bankroll
        bankroll = bet(odds, bankroll)
        bankroll_history.append(bankroll)
        
    return bankroll_history


# Odds for three candidates in the market
odds_sets = np.array([2, 3, 4])  # Example odds for 3 candidates

# Run the simulation with 100 rounds and an initial bankroll of $1000
bankroll_history = run_simulation(odds_sets, initial_bankroll=1000, n_rounds=100)

pd.Series(bankroll_history).plot(); plt.show()
```
