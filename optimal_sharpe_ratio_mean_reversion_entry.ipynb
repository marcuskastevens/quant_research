```python
from typing import Dict

import numpy as np 
import pandas as pd
from tqdm import tqdm
from matplotlib import pyplot as plt
```

```python
LOC: float = 0.00
SCALE: float = 1.00
SIGMA_SCALAR: float = 1.0
```

```python
price: pd.DataFrame = pd.DataFrame(np.random.normal(loc=LOC, scale=SCALE, size=(1_000, 1_000)))
price
```

```python
sigma: pd.DataFrame = price.ewm(halflife=10, min_periods=20).std().shift(1)
sigma
```

```python
fair_value: pd.DataFrame = price.ewm(halflife=10, min_periods=20).mean().shift(1)
fair_value
```

```python
lower_band: pd.DataFrame = fair_value - SIGMA_SCALAR * sigma
upper_band: pd.DataFrame = fair_value + SIGMA_SCALAR * sigma
```

```python
price.iloc[:200, 0].plot()
fair_value.iloc[:200, 0].plot()
lower_band.iloc[:200, 0].plot()
upper_band.iloc[:200, 0].plot()
```

```python
view: Dict = {}
trade_price: Dict = {}

for i, price_i in tqdm(price.T.items()):
        
    if any(np.isnan(lower_band.iloc[i])) or any(np.isnan(upper_band.iloc[i])) or any(np.isnan(price_i)):
        view[i] = price_i * 0
        continue
    
    if i == 0:
        view[i] = 1 * (price_i <= lower_band.iloc[i]) - 1 * (price_i >= upper_band.iloc[i])
    else:
        view[i] = (
            1 * (((price_i <= lower_band.iloc[i]) & (view.get(i-1) == 0)) | ((view.get(i-1) == 1) & (price_i <= fair_value.iloc[i])))
            - 1 * (((price_i >= upper_band.iloc[i]) & (view.get(i-1) == 0)) | ((view.get(i-1) == -1) & (price_i >= fair_value.iloc[i])))
        )
    
    trade_price[i] = (
        lower_band.iloc[i] * ((view[i] == 1) & (view.get(i-1) != 1))
        + upper_band.iloc[i] * ((view[i] == -1) & (view.get(i-1) != -1))
        + fair_value.iloc[i] * ((view[i] == 0) & (view.get(i-1) != 0))
    )
    trade_price[i] += price_i * (trade_price[i] == 0)
    
view: pd.DataFrame = pd.DataFrame(view).T
trade_price: pd.DataFrame = pd.DataFrame(trade_price).T

view
```

```python
def simulate(price: pd.DataFrame, sigma_scalar: float = 0.75) -> pd.DataFrame:

    view: Dict = {}
    trade_price: Dict = {}
    
    sigma: pd.DataFrame = price.ewm(halflife=10, min_periods=20).std().shift(1)
    fair_value: pd.DataFrame = price.ewm(halflife=10, min_periods=20).mean().shift(1)

    lower_band: pd.DataFrame = fair_value - sigma_scalar * sigma
    upper_band: pd.DataFrame = fair_value + sigma_scalar * sigma

    for i, price_i in tqdm(price.T.items()):
            
        if any(np.isnan(lower_band.iloc[i])) or any(np.isnan(upper_band.iloc[i])) or any(np.isnan(price_i)):
            view[i] = price_i * 0
            continue
        
        if i == 0:
            view[i] = 1 * (price_i <= lower_band.iloc[i]) - 1 * (price_i >= upper_band.iloc[i])
        else:
            view[i] = (
                1 * (((price_i <= lower_band.iloc[i]) & (view.get(i-1) == 0)) | ((view.get(i-1) == 1) & (price_i <= fair_value.iloc[i])))
                - 1 * (((price_i >= upper_band.iloc[i]) & (view.get(i-1) == 0)) | ((view.get(i-1) == -1) & (price_i >= fair_value.iloc[i])))
            )
        
        trade_price[i] = (
            lower_band.iloc[i] * ((view[i] == 1) & (view.get(i-1) != 1))  # long entry at the lower band
            + upper_band.iloc[i] * ((view[i] == -1) & (view.get(i-1) != -1)) # short entry at the upper band
            + fair_value.iloc[i] * ((view[i] == 0) & (view.get(i-1) != 0))  # close the trade once the price hits fair value
        )
        trade_price[i] += price_i * (trade_price[i] == 0)  # the price if there is no trading activity
        
    view: pd.DataFrame = pd.DataFrame(view).T
    trade_price: pd.DataFrame = pd.DataFrame(trade_price).T

    trade_returns = trade_price - trade_price.shift(1)
    strategy_returns = view.shift(1) * trade_returns

    cumulative_returns = strategy_returns.cumsum() 
    ending_wealth = cumulative_returns.iloc[-1].mean()

    flattened_returns = pd.Series(strategy_returns.to_numpy().flatten()).dropna()
    non_zero_flattened_returns = flattened_returns[flattened_returns != 0]

    sharpe_ratio = non_zero_flattened_returns.mean() / non_zero_flattened_returns.std(ddof=1)

    return sharpe_ratio, ending_wealth, cumulative_returns
```

# Gaussian MRP Optimization

This will explore the parameter sensitivity of a standard, normally distributed mean reverting stochastic process. 

This is a reasonable prior for any MRP and serves as a natural starting point to explore more robust approaches to non-standard MRPs.

```python
price: pd.DataFrame = pd.DataFrame(np.random.normal(loc=LOC, scale=SCALE, size=(1_000, 1_000)))
price
```

```python
sigma_scalar_to_sharpe_ratio_map = {}
sigma_scalar_to_ending_wealth_map = {}
sigma_scalar_to_cumulative_return_map = {}

for sigma_scalar in tqdm(np.arange(0.0, 5.0, step=0.10)):
    sigma_scalar_to_sharpe_ratio_map[sigma_scalar], sigma_scalar_to_ending_wealth_map[sigma_scalar], sigma_scalar_to_cumulative_return_map[sigma_scalar] = simulate(
        price=price, 
        sigma_scalar=sigma_scalar
    )
```

```python
sigma_scalar_to_sharpe_ratio_map = pd.Series(sigma_scalar_to_sharpe_ratio_map)
sigma_scalar_to_sharpe_ratio_map.plot(
    ylabel="Sharpe Ratio",
    xlabel="Sigma",
    title="What Entry Maximizes Sharpe Ratio?"
)
plt.show()
```

```python
sigma_scalar_to_ending_wealth_map = pd.Series(sigma_scalar_to_ending_wealth_map)
sigma_scalar_to_ending_wealth_map.plot(
    ylabel="Ending Wealth",
    xlabel="Sigma",
    title="What Entry Maximizes Ending Wealth?"
)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=10, ncols=5, figsize=(10, 20))

row = 0
col = 0

for i, key in enumerate(sigma_scalar_to_cumulative_return_map.keys()):
    
    ax[row, col].plot(sigma_scalar_to_cumulative_return_map[key])
    ax[row, col].set_title(f"Sigma Entry = {round(key, 2)}")

    col += 1
    if col == 5:
        col = 0
        row += 1

plt.tight_layout()
plt.show()
```

# Laplace MRP

Captures leptokyrtic, tail-driven behavior. This is likely to result in a smaller optimal sigma entry as there is a higher density of observations very close to the mean. However, there is also a higher density of observations at extreme tails, making this a slightly more realisitc and difficult stochastic process to trade.

```python
price: pd.DataFrame = pd.DataFrame(np.random.laplace(loc=LOC, scale=10, size=(1_000, 1_000)))
price
```

# Lognormal MRP

Captures right-skewed, tail-driven behavior. 

Optimally, lognormality requires asymmetric volatility bands; however, let's explore how the assumption of symmetry in the presence of asymmetric impacts the p/l.

This will serve as a skew parameter mispecification robustness test.

```python
price: pd.DataFrame = pd.DataFrame(np.random.lognormal(mean=LOC, sigma=SCALE, size=(1_000, 1_000)))
price
```
