# Experimental Exponential Decay Functions

These are potentially useful for modeling alpha decay or other market relationships.

```python
import pandas as pd
import numpy as np
```

Function 1: Decay According to SQRT-type relationship

```python
n = 252
gamma = 20
sharpes = np.array([1.3]*n)

def decay01(series, alpha=.94):
    decayed_series = []
    for i, x in enumerate(series):
        decayed_series.append((((alpha * (1 - alpha)**((n/gamma)*(1/(n-i)))))) * x)

    return decayed_series

pd.Series(decay01(sharpes, alpha=.9)).plot()
```

Function 2: Decay According to Exponential Function

```python
def decay02(series, alpha=.94):
    decayed_series = []
    for i, x in enumerate(series):
        decayed_series.append((((alpha * (1 - alpha)**(((i+1)))))) * x)

    return decayed_series

pd.Series(decay02(sharpes, alpha=.05)).plot()
```
