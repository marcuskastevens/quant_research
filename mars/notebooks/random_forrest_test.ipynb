```python
from models.regression import RandomForestRegressor
```

```python
# Generate some synthetic data
np.random.seed(69)
X = np.random.rand(100, 2)
y = 3 * X[:, 0] + np.sin(10 * X[:, 1]) + np.random.randn(100) * 0.1

# Fit Random Forest Regressor model
model = RandomForestRegressor(n_estimators=10, max_depth=5, random_state=69)
model = model.fit(X, y)

# Make predictions
y_hat = model.predict(X)

# Print the results
print("Predictions:", y_hat)

plt.plot(y_hat)
plt.plot(y)
plt.show()

plt.plot(y_hat - y)
plt.show()

print(np.mean(np.square(y_hat - y)))
```
